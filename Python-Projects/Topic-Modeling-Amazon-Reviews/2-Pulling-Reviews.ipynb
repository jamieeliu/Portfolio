# Importing packages, setting working directory, and loading file

```
import pickle
import json

#Store filename as string variables
working_directory = '/content/drive/MyDrive/APRD6343/TopicModeling'
working_file = '%s/reviews_Clothing_Shoes_and_Jewelry.json' % working_directory #%s gets replaced with working_directory
loadedjson = open(working_file, 'r')
```

# Parsing review data

```
#Load review data into a dictionary

count = 0
allreviews = {}
for aline in loadedjson:
  count += 1
  if count % 100000 == 0:
    print(count)
  areview = eval(aline)
  #Arbitrarily using iteration count as key for the dictionary
  allreviews[count] = areview

  #Print number of reviews for Steve Madden
print(len(allreviews))
```

```
from time import sleep

#Go through all reviews and pick out reviews that correspond to the matching ASINs
#These are reviews tied to Steve Madden

allSMasins = []

for data in open('%s/allSMasins.txt' % working_directory, 'r'):
  asins = data.split(',')
  for anasin in asins:
    allSMasins.append(anasin)
print(allSMasins)
```

# Export Steve Madden reviews to use in the next file (3-Preprocessing-Reviews) to preprocess the Steve Madden review data

```
SMreviews = {}
count = 0
for areview in allreviews:
  count += 1
  if count % 100000 == 0:
    print(count/5748920)
  #Setting current review as a dictionary to easily reference entries
  thereview = allreviews[areview]
  theasin = thereview['asin']
  reviewerid = thereview['reviewerID']
  rating = thereview['overall']
  if theasin in allSMasins:
    if rating > 3: #Export only reviews with a certain rating - edit for different rating, comment out for all reviews
        thekey = '%s.%s' % (theasin, reviewerid)
        SMreviews[thekey] = thereview

#Now, all Steve Madden reviews are stored in SMreviews
#How many reviews are there?
print(len(SMreviews))

#Save the data as a JSON dictionary
json.dump(SMreviews, open('%s/allSMOver3Ratingreviews.json' % working_directory, 'w'))

allSMreviews = json.load(open('%s/allSMOver3Ratingreviews.json' % working_directory, 'r'))
```
